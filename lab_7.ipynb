```python
from IPython.display import display, Math, Latex
import numpy as np
import matplotlib.pyplot as plt
from labellines import labelLine,labelLines
```

```python
display(Math('10.2x_{1}-3.2x_{2}=149'))
display(Math('-5.8x_{1}+16x_{2}=83'))
display(Math('10.3x_{1}+7.3x_{2}=274'))
```

```python
x1 = np.linspace(17.5,19)
x2 = (10.2*x1 - 149)/3.2
x3 = (83+5.8*x1)/16
x4 = (274 - 10.3*x1)/7.3
fig, (ax1, ax2, ax3) = plt.subplots(3, sharey=True, figsize=(5, 10))
fig.suptitle('Vertically stacked subplots',y=0.95)
ax1.set_title('equation 1')
ax2.set_title('equation 2')
ax3.set_title('equation 3')
ax1.plot(x1, x2, color='red')
ax2.plot(x1, x3, color='purple', linestyle='--')
ax3.plot(x1, x4, color='green', linestyle=':')
```

```python
plt.rcParams["figure.autolayout"] = True
plt.rcParams["figure.figsize"] = [8, 16]
plt.rcParams['figure.dpi'] = 100
plt.plot(x1, x2, color='orange', linestyle='-.', label='First equation')
plt.plot(x1, x3, color='teal', linestyle='--', label='Second equation')
plt.plot(x1, x4, color='navy', linestyle=':', label='Third equation')
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Linear Equations")
plt.legend()
plt.grid(color='gray', linestyle='-', linewidth=0.1)
eq12 = np.array([[10.2, -3.2], [-5.8, 16]])
coef12 = np.array([149, 83])
eq23 = np.array([[-5.8, 16], [10.3, 7.3]])
coef23 = np.array([83, 274])
eq13 = np.array([[10.2, -3.2], [10.3, 7.3]])
coef13 = np.array([149, 274])
inter12= np.linalg.solve(eq12, coef12)
inter23= np.linalg.solve(eq23, coef23)
inter13= np.linalg.solve(eq13, coef13)
x = [inter12[0],inter13[0],inter23[0]]
y = [inter12[1],inter13[1],inter23[1]]

labelLines(plt.gca().get_lines(), zorder=2.5)
plt.fill(x,y)
plt.savefig("output.png")
plt.savefig("output.jpg")
plt.savefig("output.svg")
# plt.show()
```
